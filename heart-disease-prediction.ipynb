# Importing Libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

# Data Collection and Processing

```python
df=pd.read_csv("/kaggle/input/heart-diseaseuci/heart.csv")
df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
```

```python
df.describe()
```

```python
df['target'].value_counts()
```

1-->Defective heart
0-->healthy heart

```python
x=df.drop(columns='target',axis=1)
y=df['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,stratify=y,random_state=3)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_pred,y_train)
print(training_data_accuracy)
```

```python
x_test_pred=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_pred,y_test)
print(test_data_accuracy)
```

Building a predictive system

```python
input_data=(62,0,0,140,268,0,0,160,0,3.6,0,2,2)
#change input data to numpy array
inputData=np.asarray(input_data)
#reshape the array as we are predicting for  only one instance
inputDataReshaped=inputData.reshape(1,-1)

prediction=model.predict(inputDataReshaped)
print(prediction)
if(prediction[0]==0):
    print("person doesn't have heart disease")
else:
    print('person have heart disease')
```
